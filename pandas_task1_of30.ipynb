```python
apr['hh']=apr['hh'].astype(int)
apr['mm']=apr['mm'].astype(int)
apr['ss']=apr['ss'].astype(int)
```

```python
apr['avg_res_time_insecond']=apr['hh']*3600+apr['mm']*60+apr['ss']
```
